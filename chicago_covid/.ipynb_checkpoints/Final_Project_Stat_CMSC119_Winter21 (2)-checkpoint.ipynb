## Final project for STAT/CMSC 11900 - Winter Quarter 2021 (due Friday, March 12, at noon)

This is not a group project - you should complete it individually. You may discuss the project with other students but you should design your study, plan your analyses, code, and write the report independently. You should acknowledge any help in writing. 

This project aims to analyze Chicago COVID-19 data (infections, hospitalizations, vaccination) and how the pandemic has affected or is affected by social, environmental, demographic factors. For this project, you need to decide on the question or set of questions you would like to answer. For clarification, these are few examples of questions one can investigate (and you are welcome to pursue one of these if you find them interesting):
- What are the socio-economic determinants of the vaccination administration?
- What are the socio-economic determinants of mortality from COVD-19?
- Does weather influence the evolution of infection in the city?
- Are the city and state restrictions effective for curbing the pandemic?
- What are the spatial (neighborhood, community) trends in COVID-19 infections and hospitalizations and what do they tell us about where the city will be (wrt to COVOD-19) in the summer?
- Will the university be open for regular teaching in Fall?
- Is Divvy use (bike sharing) COVID-19 safe?
- What is the impact of COVID-19 on mental health?

## The Assignment

The goal of the project is to go through the complete data science process to answer questions you have about a COVID-19 topic of your own choosing. You will acquire the data, design your visualizations, run statistical analysis, and communicate the results.

The score you receive will depend on: 
- quality of presentation (more details below)
- use and integration of data from multiple sources 
- how interesting are the questions being asked
- the quality of evidence (in data summaries, analyses etc) when drawing your conclusions

### You must submit two files:

1. The Jupyter Notebook that contains all the code you use for the analysis. 

2. A PDF report of your findings. This report should be **at most 4 pages long** (it is fine if it is shorter, but not acceptable if it is longer). Use data visualization, data summaries and statistical analyses to justify your conclusions. Make sure you indicate the source of data you use.


### Report Quality
Similarly to the projects in 11800, the quality of the writing, tables and figures is very important. Make sure that **the report does not exceed 4 pages** (including references). 

Figure should meet the following standards:
- Must be clearly labeled and referenced by those labels
- Axes should be labeled and informative
- Try to only include the most relevant plots (10 of the same kind of plot of a slight variation of the same information isn't very interesting)
- Do not include too much information in a single plot (for example, if you compute some metric for each community area in Chicago, you should not display that metric in one giant barplot with all community areas listed. We will take points off for things like this). 
- Be creative! We love good vizualizations. 


### Report Objectives 
You should formulate clearly the questions (hypotheses) you are trying to answer and the analyses you use to investigate them. **We recommend you focus on answering one question well.**

The report should cover these aspects:

-    Overview and Motivation: Provide an overview of the project goals and the motivation for it. 
-    Related Work: Anything that inspired you, such as a paper, a newspaper/magazine article etc.
-    Initial Questions: What questions are you trying to answer? How did these questions evolve over the course of the project? What new questions did you consider in the course of your analysis?
-    Data: Sources, scraping method, cleanup, etc.
-    Exploratory Analysis: What visualizations did you use to look at your data in different ways? What are the different statistical methods you considered? Justify the decisions you made. How did you reach these conclusions?
-    Final Analysis: What did you learn about the data? How did you answer the questions? How can you justify your answers?
