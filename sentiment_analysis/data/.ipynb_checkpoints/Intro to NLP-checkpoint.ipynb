```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
#label the data

data = []
data_labels = []
with open("pos_tweets.txt") as f:
    for i in f:
        data.append(i.strip())
        data_labels.append("pos")

with open("neg_tweets.txt") as f:
    for i in f:
        data.append(i.strip())
        data_labels.append("neg")
```

```python
#clean the data
def remove_user_handles():
    for i in range(len(data)):
        if "@" in data[i]:
            nonUserHandleWords = []
            words = data[i].split(" ")
            for j in range(len(words)):
                if "@" not in words[j]:
                    nonUserHandleWords.append(words[j])
            data[i] = " ".join(nonUserHandleWords)
            
def remove_numbers_from_sentences():
    for i in range(len(data)):
        if 

remove_user_handles()
```

```python
# Vectorization: the general process of turning a collection of text documents into numerical feature vectors
vectorizer = CountVectorizer(analyzer="word", lowercase=False)

features = vectorizer.fit_transform(data)
features_nd = features.toarray()
print(vectorizer.get_feature_names())
```

```python
X_train, X_test, y_train, y_test = train_test_split(features_nd, data_labels, train_size=0.80, random_state=1234)
```

```python
#Create a model
log_model = LogisticRegression()
log_model = log_model.fit(X=X_train, y=y_train)
```

```python
#Predict positive or negative for our test dataset 
y_pred = log_model.predict(X_test)
```

```python
def getPrediction(index):
    ind = features_nd.tolist().index(X_test[index].tolist())
    print("The prediction for the sentence " + data[ind] + " is " + y_pred[index] + " . The true sentiment is " + y_test[index])
```
